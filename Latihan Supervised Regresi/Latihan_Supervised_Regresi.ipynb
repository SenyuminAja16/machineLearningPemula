```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
```

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/SenyuminAja16/machineLearningPemula/refs/heads/master/playground-series-s4e5/train.csv')
train_df
```

```python
test_df = pd.read_csv('https://raw.githubusercontent.com/SenyuminAja16/machineLearningPemula/refs/heads/master/playground-series-s4e5/test.csv')
test_df
```

```python
for features in train_df.columns:
  plt.figure(figsize=(10, 6))
  sns.boxplot(x=train_df[features])
  plt.title(f'Box Plot of {features}')
  plt.show()
```

```python
q1 = train_df.quantile(0.25)
q3 = train_df.quantile(0.75)
IQR = q3 - q1

condition = ~((train_df < (q1 - 1.5 * IQR)) | (train_df > (q3 + 1.5 * IQR))).any(axis=1)
df = train_df.loc[condition, train_df.columns]
```

```python
for features in train_df.columns:
  plt.figure(figsize=(10, 6))
  sns.boxplot(x=df[features])
  plt.title(f'Box Plot of {features}')
  plt.show()
```

```python
numeric_features = df.select_dtypes(include='number').columns
numeric_features

scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
duplicates = df.duplicated()
print(df[duplicates])
```

```python
df.describe(include='all')
```

```python
num_var = df.shape[1]
n_cols = 4
n_rows = -(-num_var // n_cols)

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20, n_rows*4))
axes = axes.flatten()

for i, column in enumerate(df.columns):
  df[column].hist(ax=axes[i], bins=20, edgecolor='black')
  axes[i].set_title(column)
  axes[i].set_ylabel('value')
  axes[i].set_xlabel('Frequency')

for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
target_corr = df.corr()['FloodProbability']
target_corr_sorted = target_corr.abs().sort_values(ascending=False)

plt.figure(figsize=(10,6))
target_corr_sorted.plot(kind='bar')
plt.title('Correlation with SalePrice')
plt.xlabel('Variabels')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
x = df.drop(columns=['FloodProbability'])
y = df['FloodProbability']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(f'Jumlah Data: {len(x)}')
print(f'jumlah Data Latih: {len(x_train)}')
print(f'Jumlah Data Test: {len(x_test)}')
```

```python
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
pred_lars = lars.predict(x_test)
```

```python
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

print(f'MAE: {mae_lars}')
print(f'MSE: {mse_lars}')
print(f'R2: {r2_lars}')
```

```python
data = {
    'MAE' : [mae_lars],
    'MSE' : [mse_lars],
    'R2' : [r2_lars]
}
df_result = pd.DataFrame(data, index=['Lars'])
df_result
```

```python
LR = LinearRegression().fit(x_train, y_train)
pred_LR = LR.predict(x_test)

mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)

print(f'MAE: {mae_LR}')
print(f'MSE: {mse_LR}')
print(f'R2: {r2_LR}')
```

```python
df_result.loc['Linear Regression'] = [mae_lars, mse_lars, r2_lars]
df_result
```
